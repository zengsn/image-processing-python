# What is Pillow

Python Imaging Library (abbreviated as PIL) (in newer versions known as Pillow) is a free library for the Python programming language that adds support for opening, manipulating, and saving many different image file formats.

Some of the file formats supported are PPM, PNG, JPEG, GIF, TIFF, and BMP. It is also possible to create new file decoders to expand the library of file formats accessible.


# Usingh Pillow for imaging processing


## Read & write and display image
There are two general ways to display image, first method is using im.show() function of Pillow, however sometimes it may work well in jupyter.

The alternative method is using third party library called matplotlib to display it.

Here is an example.


```python
### read a image then display it###
from PIL import Image  
import matplotlib.pyplot as plt
img=Image.open('./street-sign.jpg')
# %matplotlib inline  #Jupyter
plt.imshow(img)
plt.show()
```

```python
#### save a image 
img=Image.open('./street-sign.jpg')
img.save('street-sign-save.jpg')
```

When you finish these steps, you will find a new file called street-sign-save.jpg in your left column 


## Transform image 
This part will tell you basic image transform information such as crop image size, rotate image, resize image etc.
### crop image
if your image is too big or some part of image is unuseful, you can crop it based on your idea.
Here is an example.

```python
### read a image then display it###
from PIL import Image  
import matplotlib.pyplot as plt
img=Image.open('./street-sign.jpg')
region = (100,100,700,450)
cropImg = img.crop(region)
plt.imshow(cropImg)
plt.show()
```

### rotate image
if some images are not in good angle, you may need to rotate them to make them look better.
Here is an example.

```python
img_rotate = img.rotate(45)  ### you can change the number of rotate angle in this function.
plt.imshow(img_rotate)
plt.show()
```

```python
img_rotate = img.transpose(Image.ROTATE_90) ### here is another method, also could rotate image in the same way. ROTATE_90 means rotate 90 degree 
plt.imshow(img_rotate)
plt.show()
```

### scale image  
sometime images are too large to process or need to cost lots of time to calculate, you can resize images to make them more samller.
Here is an example.

```python
img_resize = img.resize((400, 300)) ## Scale the image to 300*400 pixel size
plt.imshow(img_resize)
plt.show()
```

## convert image
Pillow has its own conversion () method to convert images into different formats. Take gray scale image conversion as an example:

```python
### read a image then display it###
from PIL import Image  
import matplotlib.pyplot as plt
img=Image.open('./street-sign.jpg')
img_convert = img.convert('L') ##L means grapscale modeyou can try RGB, F mode.  note if having error in this line, you may need upgrade matplotlib like this: pip install matplotlib --upgrade
plt.imshow(img_convert)
plt.show()
```

## image binaryzation

Image Binarization is the process of setting the gray value of the pixels on the image to 0 or 255, that is to say, the whole image presents obvious black-and-white effect.

In digital image processing, binary image plays a very important role. The binarization of image greatly reduces the amount of data in the image, so that the outline of the object can be highlighted.

In Pillow, we use point () method to judge each pixel of an image, and set its gray value to 0 or 255 according to its relative size with the mean value, then convert it into a binary image.

```python
WHITE, BLACK = 255, 0   #define gray constant value
img_point = img.point(lambda x: WHITE if x > 128 else BLACK)  #judge each pixel
img_binary = img_point.convert('1')  #convert to binary image
plt.imshow(img_binary)
plt.show()
```

## image filtering 
Image filtering in Python mainly deals with image smoothing, sharpening and edge enhancement. You can regard it as a filter in a mobile phone.

There are many kinds of filtering in Pillow. We take relief filtering as an example.

Firstly, the filter class in Pillow is introduced

```python
from PIL import Image,ImageFilter  
img_filter = img.filter(ImageFilter.EMBOSS) ### emboss means filter type we used, you can choose anther method based on your work.
plt.imshow(img_filter)
plt.show()
```

When change other filter such edge_enhance to find egde of objects in the image

```python
from PIL import Image,ImageFilter  
img_filter = img.filter(ImageFilter.EDGE_ENHANCE) ### find traffic sign edge better.
plt.imshow(img_filter)
plt.show()
```
